```python
import ffit as ff
import matplotlib.pyplot as plt
import numpy as np
import aplot as ap

%load_ext autoreload
%autoreload 2
```

```python
x = np.linspace(0, 100, 10000)
params = (0.1, 1.0, 1.0, 1.0)
freq = params[1]
period = 1 / freq
x = np.linspace(0, period * 1, 100)


y = ff.Cos.func(x, *params)
y += np.random.normal(0, 0.3 * params[0], len(y))
with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x, y, ".", markersize=8)
    res = ff.Cos.fit(x, y).plot().res
    ax.set(xlabel="x", ylabel="y", title="Cosine fit")
    fig.tight_layout()

    fig.savefig("cos_example.png", dpi=300)
# guess = ff.Cos.guess(x, y).plot(ax).res
# y2 = cos_func(x, *res)
# ax.plot(x, y2)
res
```

```python
params = ff.HyperbolaParam(1, 2, 0, 0.5)
x = np.linspace(-5, 5, 100)


y = ff.Hyperbola.func(x, *params) + np.random.normal(0, 1, len(y))

with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x, y, ".", markersize=8)  # markerfacecolor="none")
    ff.Hyperbola.fit(x, y).plot()
    ax.set(xlabel="x", ylabel="y", title="Hyperbola fit")
    fig.tight_layout()

    fig.savefig("hyperbola_example.png", dpi=300)
```

```python
params = ff.LorentzianParam(5, 1, 3, 2)
x = np.linspace(0, 6, 100)


y = ff.Lorentzian.func(x, *params) + np.random.normal(0, 1, len(y))

with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x, y, ".", markersize=8)  # markerfacecolor="none")
    ff.Lorentzian.fit(x, y).plot()
    ax.set(xlabel="x", ylabel="y", title="Lorentzian fit")
    fig.tight_layout()

    fig.savefig("lorentzian_example.png", dpi=300)
```

```python
params = ff.ExpParam(-3, -0.5, 3)
x = np.linspace(0, 10, 100)


y = ff.Exp.func(x, *params) + np.random.normal(0, 0.5, len(y))

with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x, y, ".", markersize=8)  # markerfacecolor="none")
    ff.Exp.fit(x, y).plot()
    ax.set(xlabel="x", ylabel="y", title="Exp fit")
    fig.tight_layout()

    fig.savefig("exp_example.png", dpi=300)
```

```python
params = ff.LineParam(1, 3)
x = np.linspace(0, 10, 100)


y = ff.Line.func(x, *params) + np.random.normal(0, 5, len(y))

with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x, y, ".", markersize=8)  # markerfacecolor="none")

    ff.Line.fit(x, y).plot()
    ax.set(xlabel="x", ylabel="y", title="Line fit")
    fig.tight_layout()

    fig.savefig("line_example.png", dpi=300)
```

```python
params = ff.LogParam(3, 0.5, 3)
x = np.linspace(1, 5, 100)


y = ff.Log.func(x, *params) + np.random.normal(0, 1, len(x))

with plt.xkcd():
    fig, ax = plt.subplots()
    ax.plot(x, y, ".", markersize=8)  # markerfacecolor="none")
    print(ff.Log.fit(x, y).plot(ax).res)
    ax.set(xlabel="x", ylabel="y", title="Log fit")
    fig.tight_layout()
    
    fig.savefig("log_example.png", dpi=300)
```
